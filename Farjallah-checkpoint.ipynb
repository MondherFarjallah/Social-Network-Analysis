Q1. Import the dataset from "./facebook.ncol", extract the Giant Component. Compute the minimum value of the number of neighbours of the nodes in the GC.

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

components = graph.components().giant()

num_neighbors = components.neighborhood_size()

min_num_neighbors = min(num_neighbors)

print("Minimum number of neighbors in the Giant Component:", min_num_neighbors)
```

Q2. Import the dataset from "./facebook.ncol", extract the Giant Component. Print the ID of the node with maximum degree of the GC (Note: argmin() is the equivalent of argmax() for minimum values).

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

components = graph.components().giant()

max_degree_node = components.vs.select(_degree=components.maxdegree())[0]
print("Node ID with maximum degree in the Giant Component:", max_degree_node.index)
```

Q3. Import the dataset from "./facebook.ncol", extract the Giant Component. Print the degree of nodes of the GC whose degree is less then 150

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

components = graph.components().giant()

nodes_with_low_degree = components.vs.select(_degree_lt=150)
for node in nodes_with_low_degree:
    print("Node ID:", node.index, "Degree:", node.degree())
```

Q5. Import the dataset from "./facebook.ncol". Plot the CCDF of the degree of the Giant Component in a lin-lin scale

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

components = graph.components().giant()
degrees = components.degree()
degree_values, degree_counts = np.unique(degrees, return_counts=True)
ccdf = 1 - np.cumsum(degree_counts) / len(degrees)

plt.plot(degree_values, ccdf, marker='o')
plt.xscale('linear')
plt.yscale('linear')
plt.xlabel('Degree')
plt.ylabel('CCDF')
plt.title('CCDF of Degree in the Giant Component')
plt.grid(True)
plt.show()
```

Q7. Import the dataset from "./facebook.ncol" and extract the GC. Delete one node with minimal degree in the Giant Component.

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

components = graph.components().giant()

min_degree_node = components.vs.select(_degree_eq=min(components.degree()))[0]

components.delete_vertices(min_degree_node)
```

Q8. Import the dataset from "./facebook.ncol" and extract the GC. Print the closeness of one node with minimum degree in the Giant Component

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

components = graph.components().giant()

min_degree_node = components.vs.select(_degree_eq=min(components.degree()))[0]

# Calculate the close node
closeness_centrality = components.closeness(vertices=min_degree_node, mode="ALL")

print("Closeness centrality of the node with minimum degree:", closeness_centrality)
```

Q9. Import the dataset from "./facebook.ncol". Plot the CCDF of the degree of the ENTIRE graph (not only the GC) in a log-log scale

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)

degrees = graph.degree()
degree_values, degree_counts = np.unique(degrees, return_counts=True)
ccdf = 1 - np.cumsum(degree_counts) / graph.vcount()

plt.loglog(degree_values, ccdf, marker='o')
plt.xlabel('Degree')
plt.ylabel('CCDF')
plt.title('CCDF of Degree for the Entire Graph')
plt.grid(True)
plt.show()
```

Q10. Import the dataset from "./facebook.ncol" and extract the GC. Print the node betweenneess (using the parameter "cutoff=2" (see https://python.igraph.org/en/stable/api/igraph.GraphBase.html#betweenness) of one node with minimum degree in the Giant Component

```python
graph = Graph.Read_Ncol("./facebook.ncol", directed=False)
components = graph.components().giant()

min_degree_node = components.vs.select(_degree_eq=min(components.degree()))[0]

betweenness = components.betweenness(vertices=min_degree_node, cutoff=2)

print("Node betweenness of the node with minimum degree:", betweenness)
```
